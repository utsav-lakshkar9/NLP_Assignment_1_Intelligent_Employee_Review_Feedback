A synset is a set of synonyms that share a common meaning or concept. Each synset represents a unique sense (or meaning) of a word, which helps differentiate between words with multiple meanings. <br>
A lemma represents the specific form of a word within a synset. It’s essentially a particular way to use a word in a given context.


```python
print(wn.synsets("happy"))
```

```python
print(wn.synsets("bank"))
```

```python
synset = wn.synsets("happy")
print('Word and Type : ' + synset[0].name())
print('The meaning of the word : ' + synset[0].definition())
print('Example : ' + str(synset[0].examples()))
```

```python
## to get synonyms for a given word
print(wn.synsets('education'))
for syn in wn.synsets('cats'):
  synonyms = []
  antonyms = []
  for lemma in syn.lemmas():
      synonyms.append(lemma.name())
      if lemma.antonyms():
          antonyms.append(lemma.antonyms()[0].name())
  print("Synonyms: ", synonyms)
  print("Antonyms: ",antonyms)
```

Basic Pre-processing

```python
statement = "Following code can be used for text preprocessing useful for various NLP applications.First we need to import nltk. For a given text, we can do sentence tokenization and word tokenization using nltk library functions. We can remove the punctuations using string library."
```

```python
#  Sentence Tokenization  - Tokenizes sentences from text
from nltk.tokenize import sent_tokenize
# Word Tokenization  - Tokenizes words in sentences
from nltk.tokenize import word_tokenize
```

```python
statement.upper()
print("Upper: " + statement.upper())
```

```python
statement=statement.lower()
print("Lower: " + statement)
```

```python
sentences = sent_tokenize(statement)
print(type(sentences), len(sentences))
for sentence in sentences:
    print(sentence)
```

```python
tokens = statement.split()
print(tokens)
```

```python
words = word_tokenize(statement)
print(words)
```

```python
vocab = set(words)
print("Vocab size: ", len(vocab))
```

```python
 # Remove punctuations
 ## punct = ["!",".",",",";"]
import string

for word in words:
    if word not in string.punctuation:
        print(word)
```

```python
only_words=[w for w in words if not w in string.punctuation]
print(only_words)
```

```python
#  Count Word Frequency

freq = nltk.FreqDist(only_words)
for key, val in freq.items():
    print(str(key) + ':' + str(val))
```

```python
#Removal of stop words from the text
from nltk.corpus import stopwords
# List of English stop words
english_stop_words=set(stopwords.words("english"))
print(english_stop_words)
print(len(english_stop_words))
```

```python
# Removal of stop words from the text
keywords=[w for w in only_words if not w in english_stop_words]
print(keywords)
```

```python
# Lemmatization : reduce words to base form
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
keywords=[w for w in keywords if w in wordnet_lemmatizer.lemmatize(w)]
print(keywords)

#  Next find the roots of the word
for w in keywords:
   lemmatized_words=wordnet_lemmatizer.lemmatize(w)
   print(wordnet_lemmatizer.lemmatize(w))
```

```python
text = "Following code can be used for text preprocessing useful for various NLP applications. First needed to import nltk. For given texts, we can do sentence tokenization and word tokenization using nltk library functions. We can remove the punctuations using string library."
# wordnet_lemmatizer = WordNetLemmatizer()
tokens = word_tokenize(text)

lemmatized_tokens = [wordnet_lemmatizer.lemmatize(word) for word in tokens]
print(lemmatized_tokens)
```

```python
lemmatized = [wordnet_lemmatizer.lemmatize(word, pos=wn.VERB) for word in tokens]
print("Lemmatized as Verbs:", ' '.join(lemmatized))
```

The most freqent words are ususally commonly occuring words, also called as English plumbing that are required for correct grammatical structure and fluency. They are not very informative and do not provide indication towards the sentiment of the text or towards the topic or genre of the text.
SO, lets try to find out the rare words or words that appear only one time in the text. <br>
The <b> hapaxes() </b> function from the Frequency Distribution class in NLTK can be used to words that appear only once in a given text or corpus.


```python
from nltk.probability import FreqDist
fdist = FreqDist(tokens)

# Print frequency distribution of words
for word, freq in fdist.items():
    print(f"{word}: {freq}")
```

```python
fdist = FreqDist(keywords)
rare_words = fdist.hapaxes()  ## only one time
print(rare_words)
```

```python
import gensim
from gensim.models import Word2Vec
```

```python
sents = ["we love natural language processing", "natural language processing is fun", "we love deep learning",
         "Langugae models came from NLP community", "deep learning is a subset of machine learning"]
```

```python
sents = [word_tokenize(sent) for sent in sents]
print(sents)
```

```python
model = Word2Vec(sentences=sents, vector_size=100, window=5, min_count=1, workers=4)
```

```python
vector_love = model.wv['love']
print(vector_love.shape)
print(vector_love)
```
